```python
""" Gera a 'TABELA_CENSO_IDSUS_2010'
    - Carrega os dados do idSUS,
    - Concatena os nomes das cidades e a UF na listaCidadesUF,
    - Cria a lista de colunas da tabela juntando as 'variáveis' do Censo e os dois 
    valores da tabela idSUS ('idsus_fx' e 'Grupo Homogêneo'),
    - Cria a TABELA_CENSO_IDSUS_2010 e estabele o nome das cidades como index,
    - Volta à tabela idSUS, cria um index do tipo 'nomeCidade_UF', exclui coluna
    com nome de cidade de UF e povoa a TABELA_CENSO_IDSUS_2010 com a nota idSUS e 
    grupo homogêneo

"""

# Carrega os dados do idSUS
idSUS_Cidades = pd.read_excel(idSUS_path,usecols='A,C:E')

# Cria o index
listaCidadesUF = []
for i in range(len(idSUS_Cidades)):
    listaCidadesUF.append(idSUS_Cidades["Município"][i] + "_" + idSUS_Cidades["UF"][i])
    
# CRIA as columns da TABELA_CENSO_IDSUS_2010 (colums do idsus + censo)
nomes_colunas = variáveis.copy()
nomes_colunas.insert(1, "idsus_fx")
nomes_colunas.insert(2, "Grupo Homogêneo")

# Cria a tabela
TABELA_CENSO_IDSUS_2010 = pd.DataFrame(index = listaCidadesUF, columns= nomes_colunas)
TABELA_CENSO_IDSUS_2010.index.name= "Cidades"

# Manipula a tabela idSUS para ter Index e deletar colunas de "Município" e "UF"
idSUS_Cidades.set_index([listaCidadesUF], inplace = True)
idSUS_Cidades.index.name = 'Cidades'
idSUS_Cidades.drop(['Município', 'UF'], axis=1)

# Povoa a tabela geral com dados do idSUS
for cidadeUF in TABELA_CENSO_IDSUS_2010.index:
    TABELA_CENSO_IDSUS_2010.at[cidadeUF, "idsus_fx"] = idSUS_Cidades.loc[cidadeUF, "idsus_fx"]
    TABELA_CENSO_IDSUS_2010.at[cidadeUF, "Grupo Homogêneo"] = idSUS_Cidades.loc[cidadeUF, "Grupo Homogêneo"]
```

```python
"""'func_extract'
    A partir do df de um Estado específico extraído da função 'exec_UF', que também 
    passa as features e a uf:
        - entra em um loop for para cada um das cidades (linhas) listadas nesse df,
        - encontra a referida cidade na TABELA_CENSO_IDSUS_2010,
        - para cada das features:
            - encontra a coluna da referida feature na TABELA_CENSO_IDSUS_2010,
            - na TABELA_CENSO_IDSUS_2010 preenche o valor encontrado na linha da cidade
            e na coluna da feature igual consta no df.

Args:
    df (dataframe): extração da planilha do Censo,
    features (lista): lista do nome das features que serão usadas na planinha igual consta
        na função 'exec_UF' e que foi copiada na lista 'variáveis'.
    uf (string): unidade federativa que vai ser recebida como parâmetro para a 'exec_UF' e
        depois passada para essa função quando for chamada.
    
Returns:
    TABELA_CENSO_IDSUS_2010 preenchida com os dados do df passados.    
"""

def func_extract(df, features, uf):
    df = df.dropna()
    df.set_index(["Cidades"], inplace= True)
    
    for cidade in df.index:
        cidadeUF = str(cidade) + '_' + str(uf)
        if cidadeUF in TABELA_CENSO_IDSUS_2010.index:
            for feature in features:
                if feature in TABELA_CENSO_IDSUS_2010.keys():
                    valor = df.loc[cidade, feature]
                    TABELA_CENSO_IDSUS_2010.at[cidadeUF, feature] = valor
```
